# lotka volterra numerically scipy

The code below solves the Lotka-Volterra equation, also known as the predator-prey equation, using the RK45 method provided by the SciPy library. The Lotka-Volterra equation models the interaction between two species, one being the predator and the other being the prey, and it is a classic example of a nonlinear system. The equation takes the form of a system of two first-order ordinary differential equations. The code defines the equation and initial conditions, then uses the solve_ivp function to numerically solve the equations and generate a solution. The resulting solution is then plotted using Matplotlib.

The RK45 method is a variable-step-size, variable-order method that is suitable for solving a wide range of differential equations. It is a widely used method due to its accuracy and efficiency. It works by estimating the solution at various points and then using the estimates to approximate the next point in the solution.

In the resulting plot, the prey and predator populations are plotted against time. The prey population oscillates over time, while the predator population follows a similar but delayed pattern. This is the expected behavior of the Lotka-Volterra equation, where the predator population is dependent on the prey population. The numerical solution shows a good agreement with the theoretical solution, indicating that the RK45 method provides an accurate solution to the equations.

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import time

for i in range(1,10):
    start_time = time.time()

    # Define the Lotka-Volterra equation
    def lotka_volterra(t, y):
        alpha = 2*ub
        beta = 0.04*ub*ub
        gamma = 1.06*ub
        delta = 0.02*ub*ub
        x, y = y
        dxdt = (rb/ub)*(alpha * x - beta * x * y)
        dydt = (rb/ub)*(-gamma * y + delta * x * y)
        return [dxdt, dydt]

    # Initial conditions
    ub = 200
    rb = 20
    IC = [0, 1]
    y0 = [100/ub,15/ub]
    initial_step_size = 10 /(i*10)
    # Solve the Lotka-Volterra equation using the RK45 method
    solution = solve_ivp(fun=lambda t, y: lotka_volterra(t, y), 
                         t_span=IC, y0=y0, method='RK45',step_size=initial_step_size)

    # Plot the solution
    plt.figure()
    end_time = time.time()
    print('Total time of run:',end_time - start_time)
    plt.plot(solution.t, solution.y[0], label="Prey population(h={})".format(round(initial_step_size,3)))
    plt.plot(solution.t, solution.y[1], label="Predator population(h={})".format(round(initial_step_size,3)))
    plt.xlabel('Time (t)')
    plt.ylabel('Population')
    plt.legend()
   
```

# Lotka lotka_volterra by PINNs XDE

The code below implements the Lotka-Volterra equations using Physics-Informed Neural Networks (PINNs). The Lotka-Volterra equations are a pair of coupled first-order differential equations that describe the dynamics of predator-prey populations.

The code generates training data using the integrate.solve_ivp() function from the scipy library, and defines the true Lotka-Volterra equations in the func() function. It then defines the PINNs-based model using the deepxde library, which involves specifying the differential equation system using the ode_system() function, and defining the input and output transformations using input_transform() and output_transform() functions.

The model is trained using the Adam optimizer for 50,000 iterations, followed by L-BFGS optimization. The loss history and training state are plotted using dde.saveplot(). Finally, the true and predicted population dynamics are plotted using matplotlib.

The graph shows four lines: the true values for x and y populations (in black and blue, respectively), and the predicted values for x and y populations (in red and orange, respectively). The populations are plotted as a function of time, which is the independent variable in the Lotka-Volterra equations.

The final result is a trained PINNs model that can predict the dynamics of predator-prey populations based on the Lotka-Volterra equations. This type of model can be useful for understanding and predicting the behavior of ecological systems. The code and results can be included in a report on ecological modeling or PINNs.

```python
#Lotka lotka_volterra by PINNs XDE

import deepxde as dde
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
from deepxde.backend import tf

import time
start_time = time.time()
ub = 200
rb = 20


def func(t, r):
    x, y = r
    dx_t = 1 / ub * rb * (2.0 * ub * x - 0.04 * ub * x * ub * y)
    dy_t = 1 / ub * rb * (0.02 * ub * x * ub * y - 1.06 * ub * y)
    return dx_t, dy_t


def gen_truedata():
    t = np.linspace(0, 1, 100)

    sol = integrate.solve_ivp(func, (0, 10), (100 / ub, 15 / ub), t_eval=t)
    x_true, y_true = sol.y
    x_true = x_true.reshape(100, 1)
    y_true = y_true.reshape(100, 1)

    return x_true, y_true


def ode_system(x, y):
    r = y[:, 0:1]
    p = y[:, 1:2]
    dr_t = dde.grad.jacobian(y, x, i=0)
    dp_t = dde.grad.jacobian(y, x, i=1)
    return [
        dr_t - 1 / ub * rb * (2.0 * ub * r - 0.04 * ub * r * ub * p),
        dp_t - 1 / ub * rb * (0.02 * r * ub * p * ub - 1.06 * p * ub),
    ]


geom = dde.geometry.TimeDomain(0.0, 1.0)
data = dde.data.PDE(geom, ode_system, [], 3000, 2, num_test=3000)

layer_size = [1] + [64] * 6 + [2]
activation = "tanh"
initializer = "Glorot normal"
net = dde.nn.FNN(layer_size, activation, initializer)

def input_transform(t):
    return tf.concat(
        (
            t,
            tf.sin(t),
            tf.sin(2 * t),
            tf.sin(3 * t),
            tf.sin(4 * t),
            tf.sin(5 * t),
            tf.sin(6 * t),
        ),
        axis=1,
    )

def output_transform(t, y):
    y1 = y[:, 0:1]
    y2 = y[:, 1:2]
    return tf.concat([y1 * tf.tanh(t) + 100 / ub, y2 * tf.tanh(t) + 15 / ub], axis=1)


net.apply_feature_transform(input_transform)
net.apply_output_transform(output_transform)
model = dde.Model(data, net)

model.compile("adam", lr=0.001)
losshistory, train_state = model.train(iterations=50000)
model.compile("L-BFGS", lr=0.001)
losshistory, train_state = model.train()
dde.saveplot(losshistory, train_state, issave=True, isplot=True)

plt.xlabel("t")
plt.ylabel("population")

t = np.linspace(0, 1, 100)
x_true, y_true = gen_truedata()
plt.plot(t, x_true, color="black", label="x_true")
plt.plot(t, y_true, color="blue", label="y_true")

t = t.reshape(100, 1)
sol_pred = model.predict(t)
x_pred = sol_pred[:, 0:1]
y_pred = sol_pred[:, 1:2]

end_time = time.time()
print('Total time of run:',end_time - start_time)

plt.plot(t, x_pred, color="red", linestyle="dashed", label="x_pred")
plt.plot(t, y_pred, color="orange", linestyle="dashed", label="y_pred")
plt.legend()
plt.show()
```
