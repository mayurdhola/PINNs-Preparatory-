# 2D poisson using numpy 

This code demonstrates a basic implementation of the finite difference method and the method of relaxation for solving the Poisson equation with hard boundary conditions.The method of relaxation is a popular iterative algorithm for solving the Poisson equation numerically. This algorithm works by repeatedly updating the values of the solution at each grid point based on the values at its neighbors, until the solution converges to a steady state.

The plot generated by the code shows the numerical solution of the 2D Poisson equation. The x and y axes correspond to the spatial coordinates of the domain, and the color scale represents the magnitude of the solution at each point in the domain.

In this specific case, the solution has a value of 1.0 along the right boundary and 0.0 along the other three boundaries. This leads to a spatially varying solution that increases gradually from 0.0 near the left boundary to 1.0 near the right boundary.

The contour plot shows regions of constant solution value, with the contours getting closer together in regions of steeper variation. In this case, the contours are evenly spaced and form straight lines, indicating a linear variation in the solution.

In short, Different approach of plotting solution has been performed here. the plot provides a visual representation of the solution of the Poisson equation and allows us to gain insight into the behavior of the solution within the domain. By examining the contour lines and color scale, we can see how the solution varies across the domain and identify any regions of particularly high or low values. This information can be useful for understanding the physical system that the Poisson equation represents and for designing numerical simulations to model that system.

```python
import numpy as np
import matplotlib.pyplot as plt
import time
for i in range(1,5,1):
    start_time = time.time()
    # Define problem parameters
    Lx = 1.0     # length of domain in x-direction
    Ly = 1.0     # length of domain in y-direction
    nx = 5*i     # number of grid points in x-direction
    ny = 5*i    # number of grid points in y-direction
    dx = Lx/(nx-1)   # grid spacing in x-direction
    dy = Ly/(ny-1)   # grid spacing in y-direction
    x = np.linspace(0, Lx, nx)   # array of x-coordinates
    y = np.linspace(0, Ly, ny)   # array of y-coordinates

    # Set up initial conditions
    u = np.zeros((ny, nx))   # initialize solution array
    u[0,:] = 0.0             # bottom boundary condition
    u[-1,:] = 0.0            # top boundary condition
    u[:,0] = 0.0             # left boundary condition
    u[:,-1] = 1.0            # right boundary condition
    tol = 1e-5               # convergence tolerance
    error = 1.0              # initialize error

    # Solve the Poisson equation using the method of relaxation
    while error > tol:
        u_old = u.copy()      # make a copy of the solution
        for j in range(1, ny-1):
            for i in range(1, nx-1):
                u[j,i] = (u[j-1,i] + u[j+1,i] + u[j,i-1] + u[j,i+1])/4.0
        error = np.max(np.abs(u - u_old))  # compute the error
    end_time = time.time()
    print(end_time-start_time)
    # Plot the solution
    plt.contourf(x, y, u, cmap='viridis')
    plt.colorbar()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("2D Poisson Equation(hx={} and hy={})".format(round(x[1]-x[0],3),round(y[1]-y[0],3)))
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time
start_time = time.time()
# Define problem parameters
Lx = 1.0     # length of domain in x-direction
Ly = 1.0     # length of domain in y-direction
nx = 101     # number of grid points in x-direction
ny = 101     # number of grid points in y-direction
dx = Lx/(nx-1)   # grid spacing in x-direction
dy = Ly/(ny-1)   # grid spacing in y-direction
x = np.linspace(0, Lx, nx)   # array of x-coordinates
y = np.linspace(0, Ly, ny)   # array of y-coordinates

# Set up initial conditions
u = np.zeros((ny, nx))   # initialize solution array
u[0,:] = 0.0             # bottom boundary condition
u[-1,:] = 0.0            # top boundary condition
u[:,0] = 0.0             # left boundary condition
u[:,-1] = 1.0            # right boundary condition
tol = 1e-5               # convergence tolerance
error = 1.0              # initialize error

# Solve the Poisson equation using the method of relaxation
while error > tol:
    u_old = u.copy()      # make a copy of the solution
    for j in range(1, ny-1):
        for i in range(1, nx-1):
            u[j,i] = (u[j-1,i] + u[j+1,i] + u[j,i-1] + u[j,i+1])/4.0
    error = np.max(np.abs(u - u_old))  # compute the error
end_time = time.time()
print(end_time-start_time)
# Plot the solution
plt.contourf(x, y, u, cmap='viridis')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title("2D Poisson Equation(hx={} and hy={})".format(round(x[1]-x[0],3),round(y[1]-y[0],3)))
plt.show()
```

# 2D poisson equation by DEEPxde

```python
import deepxde as dde
import numpy as np
#Import tf if using backend tensorflow.compat.v1
from deepxde.backend import tf
from scipy.special import gamma
import time
start_time = time.time()

alpha0 = 1.8
alpha = dde.Variable(1.5)

def fpde(x, y, int_mat):
   
    if isinstance(int_mat, (list, tuple)) and len(int_mat) == 3:
        int_mat = tf.SparseTensor(*int_mat)
        lhs = tf.sparse_tensor_dense_matmul(int_mat, y)
    else:
        lhs = tf.matmul(int_mat, y)
    lhs = lhs[:, 0]
    lhs *= -tf.exp(tf.lgamma((1 - alpha) / 2) + tf.lgamma((2 + alpha) / 2)) / (
        2 * np.pi ** 1.5
    )
    x = x[: tf.size(lhs)]
    rhs = (
        2 ** alpha0
        * gamma(2 + alpha0 / 2)
        * gamma(1 + alpha0 / 2)
        * (1 - (1 + alpha0 / 2) * tf.reduce_sum(x ** 2, axis=1))
    )
    return lhs - rhs



def func(x):
    return (1 - np.linalg.norm(x, axis=1, keepdims=True) ** 2) ** (1 + alpha0 / 2)


geom = dde.geometry.Disk([0, 0], 1)

observe_x = geom.random_points(30)
observe_y = dde.icbc.PointSetBC(observe_x, func(observe_x))

data = dde.data.FPDE(
    geom,
    fpde,
    alpha,
    observe_y,
    [8, 100],
    num_domain=64,
    anchors=observe_x,
    solution=func,
)

net = dde.nn.FNN([2] + [20] * 4 + [1], "tanh", "Glorot normal")
# Backend tensorflow.compat.v1
net.apply_output_transform(
    lambda x, y: (1 - tf.reduce_sum(x ** 2, axis=1, keepdims=True)) * y
)
# Backend paddle
# net.apply_output_transform(
#     lambda x, y: (1 - paddle.sum(x ** 2, axis=1, keepdim=True)) * y
# )

model = dde.Model(data, net)
model.compile("adam", lr=1e-3, loss_weights=[1, 100], external_trainable_variables=[alpha])
variable = dde.callbacks.VariableValue(alpha, period=1000)
losshistory, train_state = model.train(iterations=10000, callbacks=[variable])
end_time = time.time()
print(end_time-start_time)
dde.saveplot(losshistory, train_state, issave=True, isplot=True)
```

The code is solving a fractional Poisson equation on a disk-shaped domain, where the fractional derivative is defined in terms of a Riesz potential. The Riesz potential is a generalization of the Laplacian operator and is related to the fractional Laplacian operator, which is used in the 2D Poisson equation with a fractional power. Therefore, this code can be seen as a generalization of the 2D Poisson equation to the case of a fractional derivative.

The numerical method used in this code is a finite difference method. The domain is discretized into a set of points, and the fractional derivative is approximated using a finite difference scheme. The resulting system of equations is solved using a neural network.

The final output of the graph is a plot of the loss history and the convergence of the model during the training process. The loss history plot shows the evolution of the loss function over the course of the training, while the convergence plot shows the convergence of the solution to the true solution as the number of iterations increases.
