# Diffusion equation using Numpy

The solution of the diffusion equation represents the evolution of a quantity, such as temperature or concentration, over time and space. In this case, the solution is obtained using the finite difference method, which approximates the second derivative in the diffusion equation using a central difference scheme. 

The PDE being solved in the code is the 1D heat/diffusion equation:

∂u/∂t = r * ∂²u/∂x²

The numerical method used to solve the PDE is the Forward-Time Central-Space (FTCS) finite difference method. The FTCS method uses a forward difference approximation in time and a central difference approximation in space.

The final solution graph shows the evolution of the initial condition as heat is being diffused through the domain over time. The 3D plot shows the solution at each time step, where the x-axis represents the space domain, the y-axis represents time, and the z-axis represents the temperature value of the solution.

```python
import numpy as np
import matplotlib.pyplot as plt
import time
for i in range(1,5):
    s = time.time()
    # Define parameters
    L = 2  # Length of the interval
    T = 1  # Total time
    n = 5*i  # Number of grid points in space
    m = 100*i  # Number of time steps
    h = L / (n - 1)  # Grid spacing in space
    k = T / m  # Time step size
    r = k / h**2  # Constant parameter

    # Initialize solution matrix
    u = np.zeros((n, m+1))

    # Set initial condition
    u[:, 0] = np.sin(np.pi * np.linspace(-1, 1, n)) * np.exp(np.linspace(0, -1, n))

    # Set boundary conditions
    u[0, :] = 0
    u[-1, :] = 0

    # Solve the diffusion equation using the finite difference method
    for j in range(m):
        for i in range(1, n-1):
            u[i, j+1] = r * (u[i-1, j] - 2*u[i, j] + u[i+1, j]) + u[i, j]

    # Plot the solution
    x = np.linspace(-1, 1, n)
    t = np.linspace(0, T, m+1)
    X, T = np.meshgrid(x, t)

    e = time.time()
    print('Total time of run:',e - s)

    fig = plt.figure(figsize=(8, 6))
    ax = fig.gca(projection='3d')
    plt.title("Diffusion Equation(h={} and t={})".format(round(x[1]-x[0],3),round(t[1]-t[0],3)))
    ax.plot_surface(X, T, u.T)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import time
s = time.time()
# Define parameters
L = 2  # Length of the interval
T = 1  # Total time
n = 40  # Number of grid points in space
m = 1000  # Number of time steps
h = L / (n - 1)  # Grid spacing in space
k = T / m  # Time step size
r = k / h**2  # Constant parameter

# Initialize solution matrix
u = np.zeros((n, m+1))

# Set initial condition
u[:, 0] = np.sin(np.pi * np.linspace(-1, 1, n)) * np.exp(np.linspace(0, -1, n))

# Set boundary conditions
u[0, :] = 0
u[-1, :] = 0

# Solve the diffusion equation using the finite difference method
for j in range(m):
    for i in range(1, n-1):
        u[i, j+1] = r * (u[i-1, j] - 2*u[i, j] + u[i+1, j]) + u[i, j]

# Plot the solution
x = np.linspace(-1, 1, n)
t = np.linspace(0, T, m+1)
X, T = np.meshgrid(x, t)

e = time.time()
print('Total time of run:',e - s)

fig = plt.figure(figsize=(8, 6))
ax = fig.gca(projection='3d')
plt.title("Diffusion Equation(h={} and t={})".format(round(x[1]-x[0],3),round(t[1]-t[0],3)))
ax.plot_surface(X, T, u.T)
plt.show()
```

# Diffusion Equation using DEEPxde

Code describes the same process as other PDEs and The graph shows that the training loss and validation loss decrease with an increasing number of iterations, indicating that the model is improving and fitting the data better. However, the validation loss starts to level off after a certain number of iterations, while the training loss continues to decrease. This suggests that the model may be overfitting the training data and not generalizing well to new data

The 3D graph obtained from the code shows the predicted solution of the given partial differential equation (PDE) over time and space. The vertical axis represents the solution value at different points in space, while the horizontal axes represent time and space. The color of the surface in the plot indicates the magnitude of the solution value, with warmer colors indicating higher values.

```python
import deepxde as dde
import numpy as np

from deepxde.backend import tf

import time
s = time.time()

def pde(x, y):
    dy_t = dde.grad.jacobian(y, x, j=1)
    dy_xx = dde.grad.hessian(y, x, j=0)
   
    return (
        dy_t
        - dy_xx
        + tf.exp(-x[:, 1:])
        * (tf.sin(np.pi * x[:, 0:1]) - np.pi ** 2 * tf.sin(np.pi * x[:, 0:1]))
    )


def func(x):
    return np.sin(np.pi * x[:, 0:1]) * np.exp(-x[:, 1:])


geom = dde.geometry.Interval(-1, 1)
timedomain = dde.geometry.TimeDomain(0, 1)
geomtime = dde.geometry.GeometryXTime(geom, timedomain)

bc = dde.icbc.DirichletBC(geomtime, func, lambda _, on_boundary: on_boundary)
ic = dde.icbc.IC(geomtime, func, lambda _, on_initial: on_initial)
data = dde.data.TimePDE(
    geomtime,
    pde,
    [bc, ic],
    num_domain=40,
    num_boundary=20,
    num_initial=10,
    solution=func,
    num_test=10000,
)

layer_size = [2] + [32] * 3 + [1]
activation = "tanh"
initializer = "Glorot uniform"
net = dde.nn.FNN(layer_size, activation, initializer)

model = dde.Model(data, net)

model.compile("adam", lr=0.001, metrics=["l2 relative error"])
losshistory, train_state = model.train(iterations=10000)
e = time.time()
print('Total time of run:',e - s)
dde.saveplot(losshistory, train_state, issave=True, isplot=True)
```
